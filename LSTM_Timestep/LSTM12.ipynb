```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Reshape
```

```python
# USDCHFM5.csv ['DateTime','Open', 'High', 'Low', 'Close', 'Volume', 'Extra'] , USDCHF1440.csv ['DateTime','Time' ,'Open', 'High', 'Low', 'Close', 'Volume', 'Extra']
colnames = ['DateTime','Open', 'High', 'Low', 'Close', 'Volume', 'Extra']
```

```python
# USDCHFM5.csv use UTF 16 encoding Format USDCHF1440.csv use UTF 8 encoding format
df = pd.read_csv(
    'USDCHFM5.csv',
    encoding='utf-16',
    header=None,
    names=colnames
)
```

```python
# Konversi kolom DateTime ke tipe datetime, lalu jadikan index
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%Y.%m.%d %H:%M')
df.set_index('DateTime', inplace=True)

# Print bossku
print(df.head())
print(df.dtypes)
```

```python
# Jika ingin hanya mengambil kolom OHLC, bisa tambahkan volume dkk
ohlc = df[['Open','High','Low','Close']]
print(ohlc.head())
```

```python

scaler = MinMaxScaler(feature_range=(0, 1))
ohlc_scaled = scaler.fit_transform(ohlc)
print("Contoh hasil normalisasi:", ohlc_scaled[:3])

#save scaler
joblib.dump(scaler, 'scaler_ohlc_multistep.save')
```

```python
def create_multistep_dataset(data, window_size=12, steps_ahead=5):
    X, y = [], []
    for i in range(len(data) - window_size - steps_ahead):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size:i + window_size + steps_ahead])
    return np.array(X), np.array(y)
```

```python
window_size = 12     # 1 jam (jika data pakai dataset usdchfm5)
steps_ahead = 5      # prediksi 5 bar ke depan 5x5menit = 25 menit
```

```python
X, y = create_multistep_dataset(ohlc_scaled, window_size, steps_ahead)
print("X shape:", X.shape)  # (samples, 12, 4) 12 timestep dengan 4 fitur dataset OHLC gacor
print("y shape:", y.shape)  # (samples, 5, 4) 5 timestep prediksi ahead dengan4 fitur dataset OHLC gacor
```

```python
# Split train-test tanpa shuffle (biar tetap urutan waktu)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape :", X_test.shape)
print("y_test shape :", y_test.shape)
```

```python

#Model 1, 2 layer LSTM 64 Neuron

model = Sequential()
model.add(LSTM(units=64, return_sequences=True, input_shape=(window_size, 4)))
model.add(Dropout(0.2))
model.add(LSTM(units=32, return_sequences=False))
model.add(Dense(units=5 * 4))         # 5 timestep × 4 fitur (OHLC)
model.add(Reshape((5, 4)))            # (5, 4) output

# Model 2, 3 layer LSTM

# model = Sequential()
# model.add(LSTM(128, return_sequences=True, input_shape=(window_size, 4)))  # Layer 1
# model.add(Dropout(0.3))
# model.add(LSTM(64, return_sequences=True))                                 # Layer 2
# model.add(Dropout(0.3))
# model.add(LSTM(32, return_sequences=False))                                # Layer 3
# model.add(Dense(5 * 4))
# model.add(Reshape((5, 4)))

#Model 3 Bidirectional

# model = Sequential()
# model.add(Bidirectional(LSTM(64, return_sequences=True), input_shape=(window_size, 4)))
# model.add(Dropout(0.3))
# model.add(LSTM(32))
# model.add(Dense(5 * 4))
# model.add(Reshape((5, 4)))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
# Latih model
history = model.fit(
    X_train, y_train,
    epochs=13,
    batch_size=64,
    validation_split=0.1,   # 10% dari train dipakai buat validasi
    verbose=1
)
```

```python
y_pred = model.predict(X_test)
```

```python
# Flatten supaya bisa inverse_transform
y_pred_flat = y_pred.reshape(-1, 4)
y_test_flat = y_test.reshape(-1, 4)

# Inverse transform ke skala asli
y_pred_inv = scaler.inverse_transform(y_pred_flat)
y_test_inv = scaler.inverse_transform(y_test_flat)

# Kembalikan ke bentuk asli
y_pred_inv = y_pred_inv.reshape(-1, 5, 4)
y_test_inv = y_test_inv.reshape(-1, 5, 4)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

feature_names = ['Open', 'High', 'Low', 'Close']

for i, name in enumerate(feature_names):
    mae = mean_absolute_error(y_test_inv[:, :, i], y_pred_inv[:, :, i])
    rmse = np.sqrt(mean_squared_error(y_test_inv[:, :, i], y_pred_inv[:, :, i]))
    print(f"{name} — MAE: {mae:.6f}, RMSE: {rmse:.6f}")
```

```python
for t in range(5):
    close_mae = mean_absolute_error(y_test_inv[:, t, 3], y_pred_inv[:, t, 3])
    print(f"Close — Timestep t+{t+1}: MAE = {close_mae:.6f}")
```

```python
# tanyain pak surya bel bener apa salahnya

feature_names = ['Open', 'High', 'Low', 'Close']
actual_colors = ['tab:blue', 'tab:green', 'tab:green', 'tab:orange']  # Warna untuk Actual
predicted_colors = ['tab:orange', 'tab:red', 'tab:pink', 'tab:gray']  # Warna untuk Predicted

plt.figure(figsize=(14, 10))

for i, name in enumerate(feature_names):
    plt.subplot(2, 2, i+1)
    # Plot Actual dengan warna berbeda
    plt.plot(y_test_inv[:, 0, i], label=f'Actual {name}', color=actual_colors[i])
    # Plot Predicted dengan warna berbeda
    plt.plot(y_pred_inv[:, 0, i], label=f'Predicted {name}', linestyle='--', color=predicted_colors[i])
    plt.title(f'{name} — Prediksi vs Aktual (Timestep t+1)')
    plt.xlabel('Sample Index')
    plt.ylabel(f'{name} Price')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Plot Loss (MSE)
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss (MSE)')
plt.plot(history.history['val_loss'], label='Validation Loss (MSE)', linestyle='--')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot MAE
plt.figure(figsize=(10, 5))
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE', linestyle='--')
plt.title('Training vs Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
error = y_test_inv[:, 0, 3] - y_pred_inv[:, 0, 3]
plt.plot(error)
plt.title("Error per Sample (Close t+1)")
plt.xlabel("Sample Index")
plt.ylabel("Prediction Error")
plt.grid(True)
plt.show()

```

```python
plt.hist(error, bins=100, edgecolor='black')
plt.title("Distribusi Error (Close t+1)")
plt.xlabel("Error (Actual - Predicted)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
